```python
######## HERE YOU CAN JUST CHANGE en TO GET RAD!!! ####### 

##-*- coding: utf-8 -*-
#"""Exact distribution for maximum number of heads in k epochs of n coin flips

#Automatically generated by Colaboratory.

#Original file is located at
#    https://colab.research.google.com/drive/14L7ZF64lEfpHn5-1srI2dYvac1O5DgYX
#"""

import scipy.special as sc
import numpy as np

# The function below calculates the exact probability of seeing a max m heads when you do k epochs of n coin flips.
# sc.comb = binomial coefficient function; sc.hyp2f1 = hypergeometric function 2F1


def prob(m,n,k):
  return (1-sc.comb(n,m+1)*sc.hyp2f1(m+1,n+1,m+2,-1))**k - (1-sc.comb(n,m)*sc.hyp2f1(m,n+1,m+1,-1))**k

xx=180 #number of patches in the smaller image 

en=int(round((xx*2)*.3,0)) #size of test set
guessmax=80 #estimate the maximum number of correct assignments for all patches in test set

print("test size =", en)


# Probability of seeing a max of x in 25 epochs of 108 coin flips
print(1-(prob(guessmax,en,25)),"percentile")
print("WL max prob",prob(80,108,25)) #for reference
print("this distrib max prob",prob(guessmax,en,25),' ',guessmax/en)
# Checking that the probabilities sum to 1
print(sum([prob(m,en,25) for m in range(0,en+1)]))

# Calculating the mean of the distribution (as a fraction of 108)
avg = sum([m*prob(m,en,25) for m in range(0,en+1)])/float(en)
print("Average: ",avg)

# Calculating the second moment and standard deviation of the distribution:
avg2 = sum([m**2*prob(m,en,25) for m in range(0,en+1)])/float(en)**2
print("Standard deviation: ",np.sqrt(avg2-avg**2))
```
